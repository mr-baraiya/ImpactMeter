# ImpactMeter Notebook Workflow
This notebook converts IPL Cricsheet JSON files into a ball-by-ball CSV using short, readable cells.

## Step 1: Setup Paths
Define input and output paths used in the pipeline.

```python
from pathlib import Path

BASE_DIR = Path.cwd()
RAW_DIR = BASE_DIR / "data" / "raw" / "ipl_json"
OUTPUT_CSV = BASE_DIR / "data" / "processed" / "ball_by_ball.csv"

print("Base dir:", BASE_DIR)
print("Raw files:", len(list(RAW_DIR.glob("*.json"))))
print("Output:", OUTPUT_CSV)
```

## Step 2: Import Conversion Helpers
Reuse the tested parser from `scripts/json_to_csv.py` to keep notebook cells minimal.

```python
from scripts.json_to_csv import collect_json_files, parse_match_file, write_csv

json_files = collect_json_files(RAW_DIR)
print("Helper import successful")
print("JSON files found:", len(json_files))
```

## Step 3: Convert JSON to Ball-by-Ball Dataset
Parse all match files and write `data/processed/ball_by_ball.csv`.

```python
all_rows = []
for idx, file_path in enumerate(json_files, start=1):
    all_rows.extend(parse_match_file(file_path))
    if idx % 100 == 0:
        print(f"Processed {idx}/{len(json_files)}")

write_csv(all_rows, OUTPUT_CSV)
print("Done")
print("Matches:", len(json_files))
print("Rows:", len(all_rows))
print("Saved:", OUTPUT_CSV)
```

Preview shape and top rows for sanity check.
